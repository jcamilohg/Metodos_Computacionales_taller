```python
#Inicia codigo tomado de la guia
def exponencial1(x, n): 
    _exp = 0
    for i in range(0, n):
      num  = ((-1) ** i ) * x ** i
      den  = np.math.factorial(i)
      _exp = _exp + num / den 
    return _exp

def exponencial2(x, n):
  _exp = 0
  for i in range(0, n):
    num  =  x ** i
    den  = np.math.factorial(i)
    _exp = _exp + num / den 
  return 1/_exp

N = 100
x = 10
e1 = exponencial1(x = x,n = N )
e2 = exponencial2(x = x,n = N )
e_exa = np.exp(-x)

print(f'{e1}')
print(f'{e2}')
print(f'{e_exa}')

x = 10
N = np.arange(3, 20, 1)
exp1_ = np.zeros(len(N))
exp2_ = np.zeros(len(N))

for i,n in  enumerate(N):
  exp1_[i] =  exponencial1(x = x, n = n)
  exp2_[i] =  exponencial2(x = x, n = n)
    
plt.figure()
plt.plot(N, exp1_,"r-", label="Serie alternante exp(x)"   )
plt.plot(N, exp2_,"b-",label="1/exp(x)"   )
plt.legend()
plt.show()
#Finaliza codigo tomado de la guia
```
